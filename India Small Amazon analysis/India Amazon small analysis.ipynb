```python
import pandas as pd

# Load the dataset (replace 'Amazon Sale Report.csv' with the actual file path if needed)
df = pd.read_csv('Amazon Sale Report.csv')
```

```python
df
```

```python
# Drop rows with any null values
df = df.dropna()
```

```python
# Drop columns with all missing values
df = df.dropna(axis=1, how='all')
```

```python
# Checking any null values
df.isnull().sum()
```

```python
# Make column names more readable by replacing underscores with spaces and capitalizing words
df.columns = df.columns.str.replace('_', ' ').str.title()
```

```python
df
```

```python
# Basic information about the dataset
data_info = df.info()
```

```python
# Summary statistics for numerical columns
summary_stats = df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Identify the total quantity sold for each category
category_sales = df.groupby('Category')['Qty'].sum().sort_values(ascending=False)
```

```python
# Visualize the sales of each category
plt.figure(figsize=(10, 6))
category_sales.plot(kind='bar', color='cyan')
plt.title('Total Quantity Sold by Category')
plt.xlabel('Category')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filter the dataset for 'kurta' and 'Set' categories
kurta_set_sales = df[df['Category'].isin(['kurta', 'Set'])]
```

```python
# Group by city and calculate the total quantity sold for each city
top_cities_kurta_set = kurta_set_sales.groupby('Ship-City')['Qty'].sum().sort_values(ascending=False).head(8)
```

```python
# Visualize the top 8 cities
plt.figure(figsize=(10, 6))
top_cities_kurta_set.plot(kind='bar', color='orange')
plt.title('Top 8 Cities for Kurta and Set Sales')
plt.xlabel('City')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.show()

top_cities_kurta_set
```

```python
# Filter the dataset for 'kurta' category and group by city
kurta_sales = kurta_set_sales[kurta_set_sales['Category'] == 'kurta'].groupby('Ship-City')['Qty'].sum().sort_values(ascending=False).head(8)

# Filter the dataset for 'Set' category and group by city
set_sales = kurta_set_sales[kurta_set_sales['Category'] == 'Set'].groupby('Ship-City')['Qty'].sum().sort_values(ascending=False).head(8)
```

```python
# Plot bar chart for 'kurta' sales
plt.figure(figsize=(10, 6))
kurta_sales.plot(kind='bar', color='blue')
plt.title('Top 8 Cities for Kurta Sales')
plt.xlabel('City')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.show()

# Plot bar chart for 'Set' sales
plt.figure(figsize=(10, 6))
set_sales.plot(kind='bar', color='green')
plt.title('Top 8 Cities for Set Sales')
plt.xlabel('City')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filter the dataset for 'kurta' category and group by size
kurta_size_sales = kurta_set_sales[kurta_set_sales['Category'] == 'kurta'].groupby('Size')['Qty'].sum().sort_values(ascending=False).head(5)

# Filter the dataset for 'Set' category and group by size
set_size_sales = kurta_set_sales[kurta_set_sales['Category'] == 'Set'].groupby('Size')['Qty'].sum().sort_values(ascending=False).head(5)
```

```python
# Plot bar chart for 'kurta' size sales
plt.figure(figsize=(10, 6))
kurta_size_sales.plot(kind='bar', color='purple')
plt.title('Top 5 Sizes for Kurta Sales')
plt.xlabel('Size')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.show()

# Plot bar chart for 'Set' size sales
plt.figure(figsize=(10, 6))
set_size_sales.plot(kind='bar', color='red')
plt.title('Top 5 Sizes for Set Sales')
plt.xlabel('Size')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filter the dataset for Bengaluru and 'kurta' category, then group by size
bengaluru_kurta_sales = kurta_set_sales[(kurta_set_sales['Ship-City'] == 'BENGALURU') & (kurta_set_sales['Category'] == 'kurta')]
bengaluru_kurta_size_sales = bengaluru_kurta_sales.groupby('Size')['Qty'].sum().sort_values(ascending=False)
```

```python
# Plot horizontal bar chart for 'kurta' size sales in Bengaluru
plt.figure(figsize=(10, 6))
bengaluru_kurta_size_sales.plot(kind='barh', color='blue')
plt.title('Sizes Ordered for Kurta in Bengaluru')
plt.xlabel('Quantity Sold')
plt.ylabel('Size')
plt.show()
```
