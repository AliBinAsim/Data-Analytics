Renameing their Columns to more standard name

```python
df = df[['freelancer_ID', 'name', 'gender', 'age', 'country', 'language',
       'primary_skill', 'years_of_experience', 'hourly_rate (USD)', 'rating',
       'client_satisfaction']]
```

```python
df
```

```python
df = df.rename(columns={'name':'Name','gender':'Gender','age':'Age','country':'Country','language':'Language','primary_skill':'Primary_Skill', 'years_of_experience': 'Years_of_Experience','hourly_rate (USD)':'Hourly_Rate','client_satisfaction':'Client_Satisfaction','rating':'Rating'})
```

```python
df
```

Renames Gender columns to more formal names

```python
df['Gender'] = df['Gender'].str.lower().replace({
    'f': 'Female',
    'female': 'Female',
    'm': 'Male',
    'male': 'Male'
})
```

```python
df
```

```python
df['Gender'].unique()
```

Remove NA values 

```python
df.dropna()
```

```python
df.rename(columns={'Hourly_Rate':'Hourly_Rate_in_USD'})
```

```python
df['Hourly_Rate'].unique()
```

Fixing Columns data of Hourly_Rate

```python
df['Hourly_Rate'] = df['Hourly_Rate'].str.replace('USD','')
df['Hourly_Rate'] = df['Hourly_Rate'].str.replace('$','')
```

```python
df
```

```python
df['Rating'] = df['Rating'].fillna('Unknown')
```

```python
df.drop_duplicates()
```

```python
df = df.dropna()
```

```python
df
```

No Duplicates Found and saving the file

```python
df.to_csv("Cleaned_Data_final_Global_Freelancers_Raw.csv",index=False)
```
